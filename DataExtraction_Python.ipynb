```python
#setting respective paths in variables.
path_ag = "D:\Project\Capstone Project\Dataset/academic_grades_report"
#path_es = "D:\Project\Capstone Project\Dataset\enrolled_students"
path_es = "Dataset\enrolled_students\enrolled_students_cln_ug_"

#init the datetime module for years.
yr = datetime.datetime.today().year
```

### Fall Semester

```python
#For the Academic Grades reports.
# merging the files.
joined_files_ag = os.path.join(path_ag, "academic_grades_report_cln_ug_Fall*.csv")
# A list of all joined files is returned.
joined_list_ag = glob.glob(joined_files_ag)
# Finally, the files are joined.
df_ag_Fall = pd.concat(map(pd.read_csv, joined_list_ag), ignore_index=True)
#converting the column names to upper case.
columns = df_ag_Fall.columns
df_ag_Fall.columns = [x.upper() for x in columns]


#df_ag_Fall.rename(columns={'Term_Desc':'TERM_DESC'}, inplace=True)

#setting the semester in a variable.
sem = 'Fall'
#init a list consisting of all the years.
year = list(range(yr-1, yr - 19, -1))


#init a datafrtame for concatenation.
df_es_Fall = pd.DataFrame()

#init of loop for going through all the years files.
for i in year:
    df_es1 = pd.read_csv(str(path_es)+ str(sem)+ str(i) +'.csv', low_memory= False) #importing the respective years csv file.
    df_es_Fall = pd.concat([df_es_Fall, df_es1], axis=0) #vertically stacking all the csv files into the dataframe.


#Filter two column values to get fresh candidates.
df_es_Fall = df_es_Fall.loc[(df_es_Fall['CLAS_DESC']=='Freshman') & (df_es_Fall['STYP_DESC']=='First Time Freshman')]


#replacing existing value with a suitable replacement.
df_es_Fall['RGRP_CODE'].replace(to_replace=['NEW_FR', 'NEWUOPI', 'SA_NEW_1', 'SA_NEW_2', 'SA_NEW_3', 'SA_UOPI_1', 'FR_ORIENT1', 'FR_ORIENT3', 'ALLREG'], value='NEW_FRESH', inplace=True)


#merging the academic grade and enrolled student datasets.
df_grfresh_Fall = df_ag_Fall.merge(df_es_Fall, on=['PIDM', 'CURR_1_1_MAJR_DESC','TERM_DESC'], how='inner')

#inspecting the count of observations and features.
df_grfresh_Fall.shape
```

```python
#merging all the semester's datasets.
df_grfresh_allsem = pd.concat([df_grfresh_Fall, df_grfresh_Spring, df_grfresh_Summer], ignore_index = True)
```

```python
df_grfresh_allsem.head()
```

```python
#inspecting the unique values of Citizen column.
df_grfresh_allsem['CITZ_DESC'].unique()
```

```python
#filling the missing values.
df_grfresh_allsem['CITZ_DESC'].fillna(value='Unknown' ,inplace= True)
#init of a dictionary with replacement of the unique values.
student_type_dict = {'Non-Citizen' : 'International', 'Citizen': 'Domestic', 'Permanent Resident' : 'Domestic', 'Unknown' : 'Unknown'}
#mapping the replacement of the values in new column.
df_grfresh_allsem['STUD_TYPE'] = df_grfresh_allsem['CITZ_DESC'].map(student_type_dict)
#inspection after transformation.
df_grfresh_allsem['STUD_TYPE'].unique()
```

```python
#inspecting the unique values of Ethnicity column.
df_grfresh_allsem['ETHN_DESC'].unique()
```

```python
#filling the missing values.
df_grfresh_allsem['ETHN_DESC'].fillna(value='Unknown' ,inplace= True)
#replacing existing value with a suitable replacement.
df_grfresh_allsem['ETHN_DESC'].replace(to_replace='Unknown/Other', value='Unknown', inplace= True)
df_grfresh_allsem['ETHN_DESC'].replace(to_replace='2 or more races', value='Biracial', inplace= True)
df_grfresh_allsem['ETHN_DESC'].replace(to_replace=['Asian/Pacific Islander', 'Ntv Hawaiian or Pacific Islndr'], value='Native Hawaiian, and Other Pacific Islander', inplace= True)
#inspection after transformation.
df_grfresh_allsem['ETHN_DESC'].unique()
```

```python
#joining values to create a new column.
df_grfresh_allsem['CRSE_NAME_CODE'] = df_grfresh_allsem['SUBJ_CODE'].map(str) + df_grfresh_allsem['CRSE_NUMB'].map(str)
#separating colum value to two new column.
df_grfresh_allsem[['SEMESTER', 'YEAR']] = df_grfresh_allsem['TERM_DESC'].str.extract(r'(.+) (\d+)')
```

```python
#filtering to get only the features needed in the final dataframe.
df_final = df_grfresh_allsem[['PIDM','CURR_1_1_MAJR_DESC','CURR_1_DEGC_DESC_x', 'CURR_1_COLL_DESC', 'CURR_1_1_DEPT_DESC','CRSE_NAME_CODE', 'CRSE_TITLE', 'CRN', 'INSTR_1_FMIL', 'INST_GPA_x', 'TERM_INST_GPA_x','SEMESTER', 'YEAR', 'GENDER_DESC', 'ETHN_DESC', 'STUD_TYPE', 'PR_CITY', 'PR_STAT_CODE','PR_NATN_DESC']]
```

```python
#printing sample of the dataframe.
df_final.head(10)
```

```python
df_final.loc[:, 'PIDM'] = df_final.loc[:,'PIDM'].fillna(0).astype('int')
```

```python
#init a list with all the state codes of United States.
state_list = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DC", "DE", "FL", "GA", "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]
#joining the list of state codes.
pat = r'\b{}\b'.format('|'.join(state_list))
#replacing the value in the column 'PR_NATN_DESC' based on the given conditions.
df_final.loc[df_final.PR_STAT_CODE.str.contains(pat, na=False), 'PR_NATN_DESC'] = 'United States'
#replacing the value in the column 'PR_NATN_DESC' based on the other column values.
df_final.loc[df_final.STUD_TYPE == 'Domestic', 'PR_NATN_DESC'] = 'United States'
#replacing the value in the column 'STUD_TYPE' based on the column values of 'PR_NATN_DESC.
df_final.loc[df_final.PR_NATN_DESC == 'United States', 'STUD_TYPE'] = 'Domestic'
```

```python
#replacing the value in the column 'STUD_TYPE' based on the column values of 'PR_NATN_DESC.
df_final.loc[(df_final['PR_NATN_DESC']!='United States') & (pd.notnull(df_final['PR_NATN_DESC']) & (df_final['PR_NATN_DESC']!='Unknown')), 'STUD_TYPE'] = 'International'
```

```python
#replacing existing value with a suitable replacement.
df_final.loc[(df_final.PR_CITY == 'Tokyo, Japan'), 'PR_CITY'] = 'Tokyo'
#init a dict for PR_NTAN_DESC key-value pair.
state_country = {'Tokyo':'Japan', 'Port Moody':'Canada', 'Ottawa':'Canada'}
#mapping the replacement of the values in dataframe.
df_final.loc[(df_final.PR_NATN_DESC.isnull()), 'PR_NATN_DESC'] = df_final['PR_CITY'].map(state_country)
```

```python
#filling the missing values.
df_final.loc[df_final['PR_CITY'].isnull(), 'PR_CITY'] = 'Unknown'
df_final.loc[:,'PR_NATN_DESC'].fillna(value='Unknown', inplace=True)
df_final.loc[df_final['PR_STAT_CODE'].isnull(), 'PR_STAT_CODE'] = 'Unknown'
df_final.loc[df_final['INSTR_1_FMIL'].isnull(), 'INSTR_1_FMIL'] = 0
df_final.loc[df_final['INST_GPA_x'].isnull(), 'INST_GPA_x'] = 0
df_final.loc[df_final['TERM_INST_GPA_x'].isnull(), 'TERM_INST_GPA_x'] = 0
```

```python
#renaming the column features.
df_final.rename(columns={'CURR_1_1_MAJR_DESC':'DEGREE_MAJOR', 'CURR_1_DEGC_DESC_x':'CURRICULUM_DEGREE', 'CURR_1_COLL_DESC':'DEGREE_COLLEGE_NAME', 'CURR_1_1_DEPT_DESC':'MAJOR_DEPARTMENT_DESC', 'Crse_Title': 'CRSE_TITLE', 'Instr_1_FMIL':'CRSE_INSTRUCTOR',
                          'TERM_INST_GPA_x':'TERM_INST_GPA', 'INST_GPA_x':'INST_GPA'}, inplace=True)
```
